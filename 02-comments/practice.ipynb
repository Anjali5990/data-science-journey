Problem 65: Implicit vs Explicit Conversion

```python
# Part A: Implicit Conversion

x = 10
y = 5.5

result = x + y

print("Part A: Implicit Conversion")
print("Result:", result)
print("Datatype:", type(result))

# Part B: Explicit Casting
y = int(y)   # Explicitly convert float to int

result = x + y

print("\nPart B: Explicit Casting")
print("Result:", result)
print("Datatype:", type(result))
```
